```python
# visual inspection
print(txt)
```

```python
# remove special chars: . , ( )
txt = [re.sub('[.,()]','',x) for x in txt.split()]
txt = " ".join(txt)

# remove text like '[...]' for example '[a]'
txt = txt.replace("[", " ") # splits 'president[c]' in 'president' & 'c]'
txt = [x for x in txt.split() if not x.endswith("]")] # selects words without ...]
txt = " ".join(txt)
txt
```
