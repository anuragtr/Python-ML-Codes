```python
Tweets_Airline1 = pd.read_csv('C:/Users/anurag.trivedi/Dropbox/Job 06 Sep 2018/prep/x2-more new projects & new techniques/1. Techniques & Algos/10. Text Analytics/3. Sentiment Analysis/Data/Tweets_edit_Airline1.csv')
Tweets_Airline1.head()
```

```python
# converting to list for
Tweets_Airline1 = Tweets_Airline1['text'].values.tolist()
Tweets_Airline1
```

```python
# Polarity - Sentence wise

sid = SentimentIntensityAnalyzer()
for sentence in Tweets_Airline1:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in ss:
        print('{0}: {1}, '.format(k, ss[k]), end='')
    print()
# refer to 'compound' to check sentiment

# How it works:
    # https://medium.com/analytics-vidhya/simplifying-social-media-sentiment-analysis-using-vader-in-python-f9e6ec6fc52f
```

```python
### Understanding Output
# "Note: VADER collects and scores negative, neutral, and positive words and features (and accounts for factors like 
# negation along the way). The ‚Äúneg‚Äù, ‚Äúneu‚Äù, and ‚Äúpos‚Äù values describe the fraction of weighted scores that fall into 
# each category. VADER also sums all weighted scores to calculate a ‚Äúcompound‚Äù value normalized between -1 and 1; 
# this value attempts to describe the overall affect of the entire text from strongly negative (-1) to strongly positive (1). 
# In this case, the VADER analysis describes the passage as slightly-to-moderately negative (-0.3804)."
# Link: https://programminghistorian.org/en/lessons/sentiment-analysis
```

```python
### Theory
# "VADER (Valence Aware Dictionary and sEntiment Reasoner) is a sentiment intensity tool added to NLTK in 2014. 
# Unlike other techniques that require training on related text before use, VADER is ready to go for analysis 
# without any special setup. VADER is unique in that it makes fine-tuned distinctions between varying degrees of 
# positivity and negativity. For example, VADER scores ‚Äúcomfort‚Äù moderately positively and ‚Äúeuphoria‚Äù extremely positively. 
# It also attempts to capture and score textual features common in informal online text such as capitalizations, exclamation 
# points, and emoticons"
# Link: https://programminghistorian.org/en/lessons/sentiment-analysis
```

```python
# Overall Word Cloud - Generation

# convert list into string
text = ''.join(Tweets_Airline1)
# all to lowercase
text = text.lower()
# removing high freq & not useful words (after analysing wordcloud)
text = text.replace('virginamerica', '')
text = text.replace('flight', '')
text = text.replace('co', '')
text = text.replace('airline', '')

def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 21.0 / 255.0) # 21 for Orange
    s = int(100.0 * 0.0 / 255.0)
    l = int(100.0 * float(random_state.randint(10, 10)) / 255.0) # '60-120' reference; higher=brighter
    # for orange: h=21, s=255; l=med, 110-140
    # for black: s=0; l=lower side, 10-20
    # for white/whitish: s=255; l=higher side, 240-250

    return "hsl({}, {}%, {}%)".format(h, s, l)

# generate wordcloud
def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(
#                           font_path='/Library/Fonts/Verdana.ttf',
                          background_color='white',
                          relative_scaling = 0.6, # 'auto' also
                          stopwords = STOPWORDS, # or {'to', 'of'}
                          width=300, 
                          height=200,
#                           max_words=200, 
#                           min_font_size=10,
                          color_func=random_color_func, # remove this for multi-color
#                           repeat=False,
                          ).generate(text)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()

generate_wordcloud(text)
```

```python
# Positive/Negative/Neutral Word Cloud - List Creation

Tweets_Airline1_Pos = []
Tweets_Airline1_Neg = []
Tweets_Airline1_Neutral = []

sid = SentimentIntensityAnalyzer()
pos_count = 0
neg_count = 0
neutral_count = 0
for sentence in Tweets_Airline1:
    ss = sid.polarity_scores(sentence)
    if(ss['compound'] > 0):
        pos_count+=1
        Tweets_Airline1_Pos.append(sentence)
    else:
        if(ss['compound'] < 0):
            neg_count+=1
            Tweets_Airline1_Neg.append(sentence)
        else:
            neutral_count+=1
            Tweets_Airline1_Neutral.append(sentence)

# QC
print(pos_count)
print(len(Tweets_Airline1_Pos))
print(neg_count)
print(len(Tweets_Airline1_Neg))
print(neutral_count)
print(len(Tweets_Airline1_Neutral))
```
